```python
df = pd.read_excel('trefwoorden sdgs.xlsx')

sdgs = []
for i, row in df.iterrows():
    words = [word.lower() for word in row.values[2::] if isinstance(word,str)]
    sdgs.append(dict(nr=row.SDG_nr,name=row.SDG_name,words=words))
```

```python
results = dict()

for i, q in enumerate(questions):
    scores = dict()
    try:
        for sdg in sdgs:
            score = 0
            for word in sdg['words']:
                if word in q['description'].replace('\n',' ').lower():
                    score += 1
            if score > 0:
                scores[sdg['name']] = score
        if scores and 'partijen' in q:
            for sdg in scores.keys():
                if sdg not in results:
                    results[sdg] = dict()
                for partij in q['partijen']:
                    if partij not in results[sdg]:
                        results[sdg][partij] = 0
                    results[sdg][partij] += 1
    except Exception:
        pass
                
    if i % 1000 == 0:
        print(i)
```

```python
df = pd.DataFrame.from_dict(results).fillna(0)
df.a
```
